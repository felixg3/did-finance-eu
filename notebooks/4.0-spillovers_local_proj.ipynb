# Spillover Effects via Local Projections

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
horizon=20
res=pd.DataFrame({"h":range(horizon+1)})
res["beta"]=np.random.normal(size=horizon+1)
res["se"]=0.1
```

```python
res["upper"]=res.beta+1.64*res.se
res["lower"]=res.beta-1.64*res.se
res.plot(x="h",y="beta",yerr=1.64*res.se,figsize=(6,4));plt.tight_layout();
```
